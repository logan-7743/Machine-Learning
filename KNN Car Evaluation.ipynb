```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import neighbors, metrics
from sklearn.preprocessing import LabelEncoder as Le
```

```python
#KNN uniform vs distance
#Uniform gives the same level of importance to each neighbor
#Distnace gives high impoertnace to closer neighbors
```

```python
data = pd.read_csv("/content/car.csv")
print(data)
```

```python
data.columns = ["buying", "maint", "doors", "persons", "lug_boot", "safety", "class"]
print(data)
```

```python
X = data[["buying","maint","safety"]].values
y = data["class"]
print(X,y)
```

```python
# Iterate over the columns of X and transform text data to numeric
encoders = [Le() for _ in range(X.shape[1])]

# Iterate over the columns of X and transform text data to numeric
for i in range(X.shape[1]):
    X[:, i] = encoders[i].fit_transform(X[:, i])

print(X)
```

```python
#Conversoin for Y with mapping
label_maping = {
    'unacc':0,
    'acc'  :1,
    'good' :2,
    'vgood':3
}

y= y.map(label_maping)
print(y)
```

```python
knn = neighbors.KNeighborsClassifier(n_neighbors=25, weights="uniform")
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.2)
```

```python
knn.fit(X_train,y_train)
print(X_train)
```

```python
prediction = knn.predict(X_test)
```

```python
accuracy = metrics.accuracy_score(y_test, prediction)
print(accuracy)
```
